```python
from vpython import *

scene.background=vec(0,.6,.6)

#Criação dos objetos

redBall=sphere(color=color.red, pos=vec(0,13,0), radius=1.5)
floor=box(pos=vec(0,-5,0), size=vec(40,0.01,40), texture=textures.wood)

grafico=graph(xtitle="Time[s]", ytitle="Posição[m/s]", title="Posição vs. Tempo")
f1=gdots(color=color.blue)

time=0.0
speed=0.0
acc=-9.8
dt=0.01
e=.74
i=0

#loop

while (time<7.0):
  rate(1/dt)
  if(redBall.pos.y-redBall.radius < floor.pos.y + 0.5*floor.size.y):
    redBall.pos.y=floor.pos.y+0.5*floor.size.y+redBall.radius
    i+=1
    speed=-speed*e
    if(redBall.color.equals(color.red)):
      redBall.color=color.green
    else:
      redBall.color=color.red
  
  redBall.pos.y=redBall.pos.y+speed*dt
  speed=speed+acc*dt
  time=time+dt
  f1.plot(time,redBall.pos.y)
```
